Instead of Maple, I'm using an IPython notebook with sympy for symbolic calculations.  Hopefully everything is still kosher.  The symbols zeta and eta are replaced with z and h, respectively.  The sim() function simplifies the symbolic output.

```python
from sympy import *
import numpy as np

sim = simplify

init_printing()
```

Making symbols for all the variables we need

```python
x,y,z,h = symbols('x y z h')
x1,x2,x3,x4,x5,x6,x7,x8 = symbols('x1,x2,x3,x4,x5,x6,x7,x8')
y1,y2,y3,y4,y5,y6,y7,y8 = symbols('y1,y2,y3,y4,y5,y6,y7,y8')
```

Node vector in global coordinates

```python
nodes = Matrix([x1,y1,x2,y2,x3,y3,x4,y4,x5,y5,x6,y6,x7,y7,x8,y8])
```

Shape functions for Q8 element

```python
N1 = (1-z)*(1-h)*(z+h+1)/4
N2 = (1+z)*(1-h)*(z-h-1)/4
N3 = (1+z)*(1+h)*(z+h-1)/4
N4 = (1-z)*(1+h)*(z-h+1)/4
N5 = (1-z**2)*(1-h)/2
N6 = (1-h**2)*(1+z)/2
N7 = (1-z**2)*(1+h)/2
N8 = (1-h**2)*(1-z)/2
```

Building the mapping matrix from real to natural coordinates

```python
N = Matrix([[N1,0,N2,0,N3,0,N4,0,N5,0,N6,0,N7,0,N8,0],[0,N1,0,N2,0,N3,0,N4,0,N5,0,N6,0,N7,0,N8]])

x,y = N*nodes
```

x and y are functions of z and h by the shape functions for a Q8 element:

sympy.diff evaluates the derivative with respect to the given variable, eg. diff(x\*\*2) >>> 2*x, and these functions are simple enough to be accurately derived by the program.

This function gets the derivative of each element in N with respect to z or h

```python
dNz = Matrix([[sim(diff(i, z)) for i in N[0,:]], [sim(diff(i,z)) for i in N[1,:]]])
dNh = Matrix([[sim(diff(i, h)) for i in N[0,:]], [sim(diff(i,h)) for i in N[1,:]]])

dNz*nodes
```

```python
dNh*nodes
```

Make the Jacobian with the proper form by multiplying the derivatives of the shape function matrix with the nodes vector,

```python
J = Matrix([[sim((dNz*nodes)[0]), sim((dNz*nodes)[1])],[sim((dNh*nodes)[0]), sim((dNh*nodes)[1])]])
J[0,0]
```
