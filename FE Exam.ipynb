The symbols zeta and eta are represented by z and h, respectively, in the code.  The sim() function simplifies the symbolic output.

```python
from sympy import *
import numpy as np

sim = simplify

init_printing()
```

Making symbols for all the variables we need

```python
x,y,z,h = symbols('x y zeta eta')
x1,x2,x3,x4,x5,x6,x7,x8 = symbols('x1,x2,x3,x4,x5,x6,x7,x8')
y1,y2,y3,y4,y5,y6,y7,y8 = symbols('y1,y2,y3,y4,y5,y6,y7,y8')
```

Node vector in global coordinates

```python
nodes = Matrix([x1,y1,x2,y2,x3,y3,x4,y4,x5,y5,x6,y6,x7,y7,x8,y8])
xnodes = Matrix([x1,x2,x3,x4,x5,x6,x7,x8])
ynodes = Matrix([y1,y2,y3,y4,y5,y6,y7,y8])
```

Shape functions for Q8 element, reordered from the book due to the naming of the nodes in this problem.

```python
N1 = (1-z)*(1-h)*(z+h+1)/4
N2 = (1-z**2)*(1-h)/2
N3 = (1+z)*(1-h)*(z-h-1)/4
N4 = (1-h**2)*(1+z)/2
N5 = (1+z)*(1+h)*(z+h-1)/4
N6 = (1-z**2)*(1+h)/2
N7 = (1-z)*(1+h)*(z-h+1)/4
N8 = (1-h**2)*(1-z)/2
```

Building the mapping matrix from real to natural coordinates

```python
N = Matrix([[N1,0,N2,0,N3,0,N4,0,N5,0,N6,0,N7,0,N8,0],[0,N1,0,N2,0,N3,0,N4,0,N5,0,N6,0,N7,0,N8]])

x,y = N*nodes
```

Positions x and y are functions of z and h by the shape functions for a Q8 element:

```python
x
```

```python
y
```

sympy.diff evaluates the derivative with respect to the given variable, eg. diff(x\*\*2) >>> 2*x, and these functions are simple enough to be accurately derived by the program.

This function gets the derivative of each element in N with respect to z or h

```python
shapes = Matrix([N1,N2,N3,N4,N5,N6,N7,N8])
dNz = Matrix([[sim(diff(i,z)) for i in shapes]])
dNh = Matrix([[sim(diff(i,h)) for i in shapes]])

dNz
```

```python
dNh
```

Make the Jacobian with the proper form (eq 10-11) by multiplying the derivatives of the shape function matrix wrt z and h with the nodes vector,

```python
J = Matrix([[sim(dNz*xnodes), sim(dNz*ynodes)],[sim(dNh*xnodes), sim(dNh*ynodes)]])
J[0,0]
```

```python
J[0,1]
```

```python
J[1,0]
```

```python
J[1,1]
```

The determinant is way too long to print symbolically here without specifying the node coordinates.  For our problem we have coordinates (very simple ones, actually), which we can substitute by pairing the values with the nodes.

```python
n63 = [[x1, -1],[y1,-1],[x2,0],[y2,-1],[x3, 1],[y3,-1],[x4,1],[y4,0],[x5, 1],[y5,1],[x6,0],[y6,1],[x7, -1],[y7,1],[x8,-1],[y8,0]]
J = sim(J.subs(n63))
detJ = sim(J.det())[0]
detJ
```

We need the D and B matrices as well.  The D matrix is the material stiffness, which is the inverse of the compliance matrix we're given in the problem.  The leading term, separated from D for now, is $\frac{E}{(1-\nu^2)}$.

```python
nu, E = symbols('nu, E')
c = E/(1-nu**2)
D = Matrix([[1, nu, 0],[nu, 1, 0],[0, 0, (1-nu)/2]])
D
```

Now we'll get the derivatives of each shape function wrt x and y.  From the book, use (10-14) to get each pair of derivatives,

```python
dNxy = [sim(J**(-1)*Matrix([[dNz[i]],[dNh[i]]])) for i in range(8)]
dNxy
```

And form the B matrix from what we just calculated.

```python
B = Matrix([[dNxy[0][0], 0, dNxy[1][0], 0, dNxy[2][0], 0, dNxy[3][0], 0, dNxy[4][0], 0, dNxy[5][0], 0, dNxy[6][0], 0, dNxy[7][0], 0],
            [0, dNxy[0][1], 0, dNxy[1][1], 0, dNxy[2][1], 0, dNxy[3][1], 0, dNxy[4][1], 0, dNxy[5][1], 0, dNxy[6][1], 0, dNxy[7][1]],
            [dNxy[0][1], dNxy[0][0], dNxy[1][1], dNxy[1][0], dNxy[2][1], dNxy[2][0], dNxy[3][1], dNxy[3][0], dNxy[4][1], dNxy[4][0], dNxy[5][1], dNxy[5][0], dNxy[6][1], dNxy[6][0], dNxy[7][1], dNxy[7][0]]])
B
```

```python
BT = B.transpose()
```

The full stiffness matrix is then, $K^e = t\int_{-1}^1\int_{-1}^1 \mathbf{B^T D B}|\mathbf{J}|d\zeta d\eta$.

For our problem, we only need to get 3 of the elements, so we'll find those 3 elements before doing the integral.

```python
ddKe = BT*c*D*B*detJ
```

```python
ddK11 = ddKe[0,0]
ddK11
```

Now we just need to integrate twice, over $\zeta$ and over $\eta$.  Symbolically, this is too much for this program (look at all those $\zeta$s and $\eta$s), so we'll just manually do 2 point Gauss-Legendre integration.  The points are at $\pm \sqrt{3}^{-1}$ with weights 1 each.

```python
def glInt2(ddKii):
    dKii = ddKii.subs([[z, -1/np.sqrt(3)]]) + ddKii.subs([[z,1/np.sqrt(3)]])
    Kii = dKii.subs([[h, -1/np.sqrt(3)]]) + dKii.subs([[h, 1/np.sqrt(3)]])
    return Kii

K11 = glInt2(ddK11)
subEnu = [[E,200e9],[nu,.3]]
K11.subs(subEnu)
```

```python
ddK22 = ddKe[1,1]
ddK22
```

```python
K22 = glInt2(ddK22) 
K22.subs(subEnu)
```

```python
ddK12 = ddKe[0,1]
ddK12
```

```python
K12 = glInt2(ddK12)
K12.subs(subEnu)
```
